### Auxiliary methods

```python
def update_dict(src, updates=None, as_copy=False):
    '''
    Helper function for updating nested dictionaries.
    '''
    if as_copy: src = src.copy()
    if (updates is None): return src
    for key, new in updates.items():
        old = src.get(key, None)
        if isinstance(old, dict) and isinstance(new, dict):
            src[key] = update_dict(old, new, as_copy=as_copy)
        else:
            src[key] = new
    return src
```

```python
def collect_results(experiments, id_formatter, enforce_git=True):
    results, tasks = dict(), dict()
    for experiment_id, experiment in experiments.items():
        if enforce_git and experiment['meta'].get('git', {}).get('dirty', True):
            continue
        
        result_id = id_formatter(experiment) #get ID for entry
        if (result_id is None): continue

        # Get i/o pairs per simulation time
        result =\
        {
            'outputs' : np.squeeze(tuple(experiment['outputs'].values()), -1),
            'suggest_runtime' : np.mean(list(experiment['runtimes']['suggest'].values())),
        }


        num_initial = experiment['meta']['num_initial']
        ground_truths = np.squeeze(tuple(experiment['ground_truths'].values()))
        best_outputs = np.minimum.accumulate(result['outputs'])
        best_indices = np.argmin(best_outputs[:, None] != result['outputs'], -1)
        result['best_seen'] = ground_truths[best_indices] 

        # Get regret values for suggested answers
        f_min = experiment['f_min']
        regrets = []
        for t, (x, f, expectation) in experiment['answer'].items():
            regrets.append(f - f_min)
        result['regret'] = np.asarray(regrets)

        result['log_regret'] = np.log10(np.maximum(1e-6, result['regret']))
        result['log_regret_obs'] = np.log10(np.maximum(1e-6, result['best_seen'] - f_min))

        if isinstance(result_id, str):
            results[result_id] = result
        else:
            sub_results = results
            for term in result_id[:-1]:
                sub_results = sub_results.setdefault(term, dict())
            sub_results[result_id[-1]] = result

    return results
```

```python
def get_style(parallelism, num_evals, style=None, as_copy=True, num_markers=5, **kwargs):
    if (style is None): style = dict()
    markevery = num_evals//(5*parallelism)
    line_alpha = 0.5
    
    light_green = (0.2, 0.8, 0.2)
#     light_orange = (1.0, 0.65, 0.25)
    light_orange = (1.0, 0.7, 0.3)
    dark_orange = (0.95, 0.35, 0.1)
    dark_blue = (0.2, 0.4, 0.7)
    
    formats =\
    {
#         'loss': 
#         {
#             'confidence_bound' : {'line_kwargs' : {'linestyle':':', 'marker':'^', 'markersize':5, 'markeredgewidth':0.6, 'markevery':  markevery}},
#             'negative_ei' : {'line_kwargs' : {'linestyle':'-', 'marker':'.', 'markersize':8, 'markeredgewidth':0.6, 'markevery':  markevery}},
#         },
        
        'use_greedy': 
        {
            1 : {'line_kwargs' : {'linestyle':':', 'marker':'^', 'markersize':5, 'markeredgewidth':0.6, 'markevery':  markevery}},
            0 : {'line_kwargs' : {'linestyle':'-', 'marker':'.', 'markersize':8, 'markeredgewidth':0.6, 'markevery':  markevery}},
        },

        #Query optimization subroutine
        'subroutine' :
        {
            'random' : {'color': light_green + (0.5,), 'line_kwargs':{'alpha':None}},
            'sgd' : {'color': light_orange + (0.5,), 'line_kwargs':{'alpha':None}},
            'cmaes' : {'color': matplotlib.colors.to_rgb('skyblue') + (0.5,), 'line_kwargs':{'alpha':None}},
#             'scipy' : {'color': matplotlib.colors.to_rgb('tab:purple') + (0.5,), 'line_kwargs':{'alpha':None}},
            'scipy' : {'color': matplotlib.colors.to_rgb('tab:pink') + (0.5,), 'line_kwargs':{'alpha':None}},
            #'nlopt' : {'color': matplotlib.colors.to_rgb('tab:purple') + (0.5,), 'line_kwargs':{'alpha':None}},
            #'halving' : {'color': matplotlib.colors.to_rgb('tab:pink') + (0.5,), 'line_kwargs':{'alpha':None}},
        },
    }

    label = {}
    order = ['subroutine', 'use_greedy', 'loss']
    for key in order:
        val = kwargs.get(key, None)
        if (val is None): continue
            
        if key in formats:
            style = update_dict(style, formats[key].get(val, None), as_copy=as_copy)

        if key == 'loss':
            label[0] = glossary['loss'][val][0] + ' w/'

        if key == 'subroutine':
            label[1] = glossary['subroutine'][val]

        if key == 'use_greedy' and val:
            label[2] = '$^*$'

            rgba = matplotlib.colors.to_rgba(style['color'])
            if style['color'][:3] in [light_orange]:
                style['color'] = dark_orange + (rgba[-1],)
            elif style['color'][:3] in [matplotlib.colors.to_rgb('skyblue')]:
                style['color'] = dark_blue + (rgba[-1],)
            elif style['color'][:3] in [matplotlib.colors.to_rgb('tab:pink')]: 
                 style['color'] = matplotlib.colors.to_rgb('tab:purple') + (rgba[-1],)
            else:
                style['color'] = [0.6*channel for channel in rgba[:-1]] + [rgba[-1]]

            
    rgb = matplotlib.colors.to_rgb(style['color'])
    style['line_kwargs']['markeredgecolor'] = matplotlib.colors.to_rgb('black') + (0.8,)
    style['line_kwargs']['markerfacecolor'] = rgb + (0.8,)

    if len(label):
        style['label'] = ''.join([v for (k,v) in sorted(label.items(), key=lambda kv:kv[0])])

    return style
```

```python
def enumerate_sorted(_dict, key=lambda kv: kv[0], **kwargs):
    iterator = enumerate(sorted(_dict.items(), key=key, **kwargs))
    return iterator

    
def fill_between(ax, xvals, yvals, bound, line_kwargs=None,
    fill_kwargs=None, **kwargs):
    if (line_kwargs is None): line_kwargs = dict()
    if (fill_kwargs is None): fill_kwargs = dict()

    line = ax.plot(xvals, yvals, **{**kwargs, **line_kwargs})
    fill = ax.fill_between(xvals, yvals + bound, yvals - bound,
                                                **{**kwargs, **fill_kwargs})
    return [line, fill]


def plot_result(ax, results, parallelism=1, num_initial=3, label=None,
    color=None, line_kwargs=None, fill_kwargs=None, use_observed=False,
    cutoff=None, **kwargs):
    if (num_initial is None): num_initial = parallelism
    if (line_kwargs is None): line_kwargs = dict()
    if (fill_kwargs is None): fill_kwargs = dict()
    
    line_kwargs = {'alpha':0.6, 'zorder':2, 'linewidth':1.1, 'label':label, **line_kwargs}
    fill_kwargs = {'alpha':0.15, 'zorder':1, **fill_kwargs}
    
    if not use_observed:
        log_regrets = np.stack([r['log_regret'] for r in results.values()])
    else:
        log_regrets = np.stack([r['log_regret_obs'] for r in results.values()])[:, num_initial-1:]
        if parallelism > 1:
            log_regrets = np.hstack\
            ([
                log_regrets[:, 0 : -1 : parallelism],
                log_regrets[:, -1:],
            ])
    
        # log_regrets = np.stack([r['log_regret_obs'] for r in results.values()])
        
    num_evals = len(tuple(results.values())[0]['outputs'])
    xvals = list(range(num_initial, num_evals, parallelism)) + [num_evals]
    if len(xvals) != log_regrets.shape[-1]: bp()
    
    num_trials = len(log_regrets)
    averages = np.nanmean(log_regrets, axis=0)
    stderrs = np.nanstd(log_regrets, axis=0)/np.sqrt(num_trials)

    if cutoff is not None:
        averages = averages[:cutoff]
        stderrs = stderrs[:cutoff]
        xvals = xvals[:cutoff]

    handles = fill_between(ax, xvals, averages, stderrs, color=color,
                            line_kwargs=line_kwargs, fill_kwargs=fill_kwargs)

    return handles, log_regrets
```

```python
def plot_results_applied(results, num_initial=3, use_observed=False, save=None):
    nrows = len(results)
    ncols = max([len(result) for result in results.values()])
#     figsize = (4 * ncols, 2 * nrows)
    figsize = (4 * ncols, 2 * nrows)

    subroutine_order =\
    {
        'random' : 0,
        'nlopt' : 1,
        'cmaes' : 2,
        'scipy' : 3,
        'sgd' : 4,
    }
    
    fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=figsize)
    if nrows == 1: axes = np.atleast_2d(axes)
    elif ncols == 1: axes = np.atleast_2d(axes).T        

    eval_counts = dict()
    for i, (task, results_i) in enumerate(results.items()):
        axes[i, 0].set_ylabel('{}'.format(glossary['task'].get(task, task)), fontsize=10)
        for j, ((input_dim, parallelism, num_evals, log2), results_ij) in enumerate_sorted(results_i):
            eval_counts[i, j] = num_evals
         
            #if i == 0:
            axes[0, j].set_title\
            (
                # "Budget for RS: $|\mathbf{{X}}|=2^{{{:d}}}$"
#                 "Equivalent budget $N{=}2^{{{:d}}}$".format({4:12,8:14,16:16}[parallelism]),
                "Equivalent budget $N=2^{{{:d}}}$".format(log2),
                fontsize=8,
            )

            if i == nrows - 1:
                axes[i, j].set_xlabel('Num. evaluations', fontsize=10)

            notes =\
            [
                [(0.065, 0.065), '$d{{=}}{:d}$'.format(input_dim)],
                [(0.170, 0.065), '$q{{=}}{:d}$'.format(parallelism)],
            ]
            for pos, text in notes:
                plt.text(*pos, text, fontsize=6, ha='center', va='center', transform=axes[i,j].transAxes)
    
            for k, (subroutine, results_ijk) in enumerate_sorted(results_ij, key=lambda kv: subroutine_order[kv[0]]):
                for l, (loss, results_ijkl) in enumerate_sorted(results_ijk):
                    for m, (use_greedy, results_ijklm) in enumerate_sorted(results_ijkl):
                        style = get_style(parallelism, num_evals, task=task,
                                          loss=loss, subroutine=subroutine,
                                          use_greedy=use_greedy)

                        _ = plot_result(axes[i, j], results_ijklm, parallelism=parallelism,
                                        use_observed=use_observed, num_initial=num_initial,
                                        **style)



    
    # Reorder s.t. EI is short before UCB in the legend (even though it gets plotted after)
    handles, labels = axes[0, 0].get_legend_handles_labels()
#     indices = np.reshape(np.arange(len(handles)), [-1, 2])
#     indices = np.ravel(np.hstack([indices[1::2][:,None], indices[::2][:,None]]))
#     handles, labels = map(lambda x: np.asarray(x)[indices].tolist(), [handles, labels])
    legend = plt.legend(handles, labels, loc='best', ncol=6, 
                        fontsize=6, frameon=False, borderpad=0,
                        bbox_to_anchor=(0.4, -0.35))

    for i, (row, task) in enumerate(zip(axes, sorted(results.keys()))):
        for j, ax in enumerate(row):    
            # ax.relim()
            # ax.autoscale_view()

            num_evals = eval_counts.get((i, j), 1024)
            xticks = [num_initial] + list(range(num_evals//4, num_evals+1, num_evals//4))
                        
            ax.set_xlim([xticks[0], xticks[-1]])
            ax.set_xticks(xticks)
            ax.set_xticks([num_initial, (num_initial + num_evals//4)/2] + list(range(num_evals//4, num_evals+1, num_evals//8)), minor=True)
            # if i+1 < len(axes): plt.setp(ax.get_xticklabels(), visible=False)

            ax.set_yticks(np.linspace(*ax.get_ylim(), 5))
            ax.yaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%.2f'))
            ax.set_yticks(np.linspace(*ax.get_ylim(), 9), minor=True)
            #if j != 0: plt.setp(ax.get_yticklabels(), visible=False)

            ax.grid(color='silver', linestyle='-', which='minor', linewidth=0.5, alpha=0.4, zorder=1)
        
    if (save is not None):
        _ = plt.savefig(save, bbox_inches='tight', dpi=256)
        
    
    return fig, axes
```

```python
def id_formatter(entry):
    '''
    Gernerate a ID for a given entry. If a list/tuple is
    returned, a nesting structure will be produced.
    '''
    entry_id =\
    [ 
        entry['meta']['loss'],
#         entry['meta']['time_limit'],
        entry['meta']['parallelism'],
        entry['meta']['subroutine'],
        entry['meta']['use_greedy'],
        entry['meta']['seed'],
    ]

    if entry['meta']['task'] in ['rfgp']:
        entry_id.insert(0, 'rfgp{:d}'.format(entry['meta']['input_dim']))
    else:
        entry_id.insert(0, entry['meta']['task'])

    return entry_id
```

```python
def id_formatter_synth(entry):
    '''
    Gernerate a ID for a given entry. If a list/tuple is
    returned, a nesting structure will be produced.
    '''
    entry_id =\
    [
        (entry['meta']['input_dim'], entry['meta']['parallelism'], entry['meta']['job_limit']),
        entry['meta']['subroutine'],
        entry['meta']['loss'],
        entry['meta']['use_greedy'],
        entry['meta']['seed'],
    ]
    
    if entry['meta']['lenscale'] == 1.0 and entry['meta']['input_dim'] != 16:
        print(entry['meta']['subroutine'], entry['meta']['input_dim'])

    if 'long' in entry['meta']['schedule']:
        entry_id.insert(1, 16)
    elif 'mid' in entry['meta']['schedule']:
        entry_id.insert(1, 14)
    elif 'short' in entry['meta']['schedule']:
        entry_id.insert(1, 12)
    else:
        raise Exception('Unknown schedule')


    return entry_id
```
